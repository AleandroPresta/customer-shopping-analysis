# Customer Shopping Trends Analysis

## Introduction

This analysis explores customer shopping behavior using a dataset of 3,336 records containing information about purchases including customer demographics, item details, purchase amounts, and shopping preferences. The dataset includes features such as age, gender, item categories, prices, locations, sizes, seasons, and payment methods.

## Data Exploration
### Importing the dataset

```python
import pandas as pd

def load_shopping_data(data_path: str):
    return pd.read_csv(data_path)
```

```python
DATA_PATH = '../data/shopping_trends.csv'
shopping_data = load_shopping_data(DATA_PATH)
shopping_data.head()
```

### Preliminary Exploration

```python
shopping_data.info()
```

Inside our dataset we have 19 columns:
1. **Customer ID**: A unique identifier for each customer (integer).
2. **Age**: The age of the customer (integer).
3. **Gender**: The gender of the customer.
4. **Item Purchased**: The name or identifier of the item purchased.
5. **Category**: The category to which the purchased item belongs.
6. **Purchase Amount (USD)**: The amount spent on the purchase in US dollars (integer).
7. **Location**: The location where the purchase was made.
8. **Size**: The size of the purchased item.
9. **Color**: The color of the purchased item.
10. **Season**: The season during which the purchase was made.
11. **Review Rating**: The rating given by the customer for the purchase (float).
12. **Subscription Status**: Indicates whether the customer has a subscription.
13. **Payment Method**: The method used for payment.
14. **Shipping Type**: The type of shipping chosen for the purchase.
15. **Discount Applied**: Indicates whether a discount was applied to the purchase. 
16. **Promo Code Used**: Indicates whether a promo code was used 
17. **Previous Purchases**: The number of previous purchases made by the customer (integer).
18. **Preferred Payment Method**: The customer's preferred method of payment.
19. **Frequency of Purchases**: How often the customer makes purchases.

#### Category Exploration
For each categorical column we are gonna explore which values are available and how many rows have a certain categorical value. 

We start with the gender of the purchaser:

```python
gender_counts = shopping_data['Gender'].value_counts()
gender_counts
```

The number of purchases made by male customers almost doubles the number of purchases made by female customers. We can see this better using a pie chart:

```python
import matplotlib.pyplot as plt

gender_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'], legend=True)
plt.ylabel('')
plt.title('Gender Distribution')
plt.show()
```

Next we are gonna explore the items purchased, using a histogram:

```python
import seaborn as sns

def plot_item_distribution(column_name: str):
    item_count = shopping_data[column_name].value_counts()
    plt.figure(figsize=(12, 8))
    sns.barplot(x=item_count.values, y=item_count.index, order=item_count.index)
    plt.xlabel('Number of Purchases')
    plt.ylabel(column_name)
    plt.show()

plot_item_distribution('Item Purchased')
```

The most purchased items are Blouses, Pants and Jewelry.

```python
plot_item_distribution('Category')
```

This category has only 4 possible values and most of the items purchased are clothes.

```python
plot_item_distribution('Location')
```

Plotting the `Location` of the purchases shows us that all of them are made in the US. The clients are spread all over the state but the states with the highest number of purchases are `Montana`, `California` and `Idaho`.

```python
plot_item_distribution('Size')
```

The most common size is M followed by L, S and XL.
